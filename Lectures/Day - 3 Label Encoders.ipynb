```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler, MinMaxScaler
```

```python
df = pd.read_csv('Pokemon.csv')
```

```python
df.info()
```

```python
df.isnull().sum()

df.dropna(inplace=True)
```

```python
df.info()
```

```python
hot = OneHotEncoder()

encoded = hot.fit_transform(df[['Type 2']])

encoded_df = pd.DataFrame(encoded.toarray(), columns=hot.get_feature_names_out(['Type 2']))

df_encoded = pd.concat([df, encoded_df], axis=1)

df_encoded.head()
```
